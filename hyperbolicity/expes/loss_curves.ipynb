```python
import re
import torch
import matplotlib.pyplot as plt
file_path = '/share/home/houedry/projects/DifferentiableHyperbolicity/hyperbolicity/expes/launch_files/results_expes/D_cbmc_2025-05-09_17-07-53/lr_0.01_dr_0.001_sd_1.0_epoch_1000_batch_32_n_batches_100.pt'
# Load the .pt file
loaded_data = torch.load(file_path)

# Extract losses from the loaded data
losses = loaded_data['losses']
match = re.search(r"lr_([0-9.]+)_dr_([0-9.]+)_sd_([0-9.]+)", file_path)
if match:
    dr = match.group(2)
    sd = match.group(3)
    lr = match.group(1)


title = f"Training Loss Over Epochs (lr={lr}, dr={dr}, sd={sd})"
# Plot the losses
plt.figure(figsize=(10, 6))
plt.plot(losses, label="Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title(title)
plt.legend()
plt.grid()
plt.show()
```
